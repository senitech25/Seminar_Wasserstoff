```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('S2_Strombilanz_monatlich.csv')

# Display the first 5 rows of the data
df.head(20)
```

```python
df['Bilanz']=df['Bilanz']/1000
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 5))

# create a new column to indicate whether the Bilanz is positive or negative
df['color'] = df['Bilanz'].apply(lambda x: '#77DD77' if x >= 0 else '#FF6961')

# create the bar plot
bars=plt.bar(df['Monat'], df['Bilanz'], color=df['color'])

# add a horizontal line at y=0
plt.axhline(y=0, color='black', linewidth=1)

# set the x-label and y-label
plt.xlabel('Monat')
plt.title('monatliche Strombilanz [MWh]')

# set the x-tick labels to be rotated by 60 degrees
plt.xticks(rotation=60)

# show the plot
plt.show()
```
