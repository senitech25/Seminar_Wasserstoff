# Stündliche Verteilung des Strombedarfs der Uni Augburg

### Using the hourly weights of the electrical energy profile from Biechele, the hourly power demands of the Uni Augsburg was determined. Here is the visual diagram of the distribution

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('uni_augsburg_hourly_demand.csv')

# Display the first 5 rows of the data
df.head(20)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 5))

# plot the line graph
ax.plot(df['Stunde'], df['Stündlicher Stromverbrauchan der Uni Augbsurg [kWh]'], color='purple', marker='o')

# plot the bar chart
ax.bar(df['Stunde'], df['Stündlicher Stromverbrauchan der Uni Augbsurg [kWh]'], color='#046563', alpha=0.5)

# set the labels and title
ax.set_xticks(df['Stunde'])
ax.set_xlabel('Stunde', fontsize=11)
#ax.set_ylabel('kWh/m2',fontsize=16)
ax.set_title('Strombedarf an der Uni [kWh]',fontsize=12)

# display the plot
plt.show()
```
